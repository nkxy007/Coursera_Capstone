### Explore all the Borough and Neighbourhoods in toronto_hoods_df_clean_TR DF
This DF contains Borough in whose names contain Toronto

Create a function to repeat the process of extracting venues in toronto_hoods_df_clean_TR

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500, LIMIT=100):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
#Data frame used toronto_hoods_df_clean_TR
names = toronto_hoods_df_clean_TR['Neighbourhood'].to_list()
latitudes = toronto_hoods_df_clean_TR['Latitude'].to_list()
longitudes  = toronto_hoods_df_clean_TR['Longitude'].to_list()
toronto_hoods_venues = getNearbyVenues(names, latitudes, longitudes, radius=500)
```

### Analyze Each Neighborshood

```python
# Create a DF of Neighborhood names and corresponding venues
toronto_hoods_venues_onehot = pd.get_dummies(toronto_hoods_venues[['Venue Category']],prefix="", prefix_sep="")
toronto_hoods_venues_onehot=toronto_hoods_venues_onehot.drop('Neighborhood', axis=1)
toronto_hoods_venues_onehot['Neighborhood']=toronto_hoods_venues['Neighborhood']
# bring the neighborhood name column at First position
fixed_columns = [toronto_hoods_venues_onehot.columns[-1]] + list(toronto_hoods_venues_onehot.columns[:-1])
toronto_hoods_venues_onehot = toronto_hoods_venues_onehot[fixed_columns]

toronto_hoods_venues_onehot.head()
```

```python
toronto_hoods_venues_onehot.shape
```

Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
toronto_hoods_venues_grouped = toronto_hoods_venues_onehot.groupby('Neighborhood').mean().reset_index()
toronto_hoods_venues_grouped
```

Let's print each neighborhood along with the top 5 most common venues

```python
num_top_venues = 5

for hood in toronto_hoods_venues_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_hoods_venues_grouped[toronto_hoods_venues_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
    
```

Let's put that into a pandas dataframe

First, let's write a function to sort the venues in descending order.


```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Now let's create the new dataframe and display the top 10 venues for each neighborhood.

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for num in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(num+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(num+1))

# create a new dataframe
toronto_hoods_venues_sorted = pd.DataFrame(columns=columns)
toronto_hoods_venues_sorted['Neighborhood'] = toronto_hoods_venues_grouped['Neighborhood']

for index in np.arange(toronto_hoods_venues_grouped.shape[0]):
    toronto_hoods_venues_sorted.iloc[index, 1:] = return_most_common_venues(toronto_hoods_venues_grouped.iloc[index, :], num_top_venues)

toronto_hoods_venues_sorted.head()
```

## Clustering Neighborhoods

 Lets device the neighborhoods into 5 clusters

```python
# set number of clusters
kclusters = 5

toronto_hoods_venues_grouped_clustering = toronto_hoods_venues_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_hoods_venues_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

Let's create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

```python
#rename toronto_hoods_df_clean_TR Neighbourhood to Neighborhood
toronto_hoods_df_clean_TR.rename(columns={'Neighbourhood':'Neighborhood'}, inplace=True)
toronto_hoods_df_clean_TR.head(2)
```

```python
# add clustering labels
toronto_hoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
toronto_hoods_venues_merged = toronto_hoods_df_clean_TR
# merge toronto_hoods_venues_sorted with toronto_hoods_df_clean_TR to add latitude/longitude for each neighborhood
toronto_hoods_venues_merged = toronto_hoods_venues_merged.join(toronto_hoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_hoods_venues_merged.head()
```

### Visualize the clusters on the map

```python
# create map centered at Parkwoods 	43.753259 	-79.329656
latitude = 43.753259
longitude = -79.329656
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_hoods_venues_merged['Latitude'], toronto_hoods_venues_merged['Longitude'], toronto_hoods_venues_merged['Neighborhood'], toronto_hoods_venues_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
